**1) Realiza una función que devuelva el área de un círculo a partir de un radio. La función se llamará area_circulo(). El radio se pedirá al usuario por teclado:**

Nota: El área de un círculo se obtiene al elevar el radio a dos y multiplicando el resultado por el número pi. Puedes utilizar el valor 3.14159 como pi o importarlo del módulo math:
```python
import math
print(math.pi)
> 3.1415...
```

````python
import math

def area_circulo():
    """
    Función que calcula el área de un círculo.
    Solicita el radio al usuario y devuelve el área.
    """

    radio = float(input("Ingresa el radio del círculo: "))
    
    area = math.pi * (radio ** 2)
    return area

resultado = area_circulo()
print(f"El área del círculo es: {resultado:.2f} unidades cuadradas")
````

**2) Realiza una función llamada lee_numero() que solicite y lea por teclado un numero. Utilizar esta función para pedirle al usuario 3 números. Luego pasarle estos 3 números a una función que se llame mayor() que tenga 3 parámetros y que devuelva el número mayor de los 3:**

````python
def lee_numero():
    """
    Función que solicita y lee un número por teclado.
    Devuelve el número ingresado convertido a float.
    """
    numero = float(input("Ingresa un número: "))
    return numero

def mayor(num1, num2, num3):
    """
    Función que recibe 3 números como parámetros.
    Devuelve el número mayor de los tres.
    """
    numero_mayor = max(num1, num2, num3)
    return numero_mayor

print("Por favor, ingresa 3 números:\n")
numero1 = lee_numero()
numero2 = lee_numero()
numero3 = lee_numero()

resultado_mayor = mayor(numero1, numero2, numero3)

print(f"\nLos números ingresados son: {numero1}, {numero2}, {numero3}")
print(f"El número mayor es: {resultado_mayor}")
````
